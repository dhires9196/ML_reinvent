```python
### training our model
#seperate our features
x=df.drop('test_result',axis=1)
```

```python
y=df['test_result']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.10, random_state=42)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaled_x_train=scaler.fit_transform(x_train)
```

```python
scaled_x_test=scaler.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model=LogisticRegression()
```

```python
log_model.fit(scaled_x_train,y_train)
```

```python
log_model.coef_
```

```python
predicted_y=log_model.predict(scaled_x_test)
```

```python
predicted_y
```

```python
predicted_y[0]
```

```python
predicted_y_prob=log_model.predict_proba(scaled_x_test)
```

```python
predicted_y_prob
```

```python
log_model.classes_
```

##### classification metrics

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,predicted_y)
```

```python
confusion_matrix(y_test,predicted_y)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test,predicted_y),display_labels=log_model.classes_).plot()
```

```python
print(classification_report(y_test,predicted_y))
```

```python
#another way for AUC/ROC curve
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt
import numpy as np
from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(y_test, predicted_y)
roc_auc = metrics.auc(fpr, tpr)
display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc,estimator_name='example estimator')
display.plot()
plt.show()
```
